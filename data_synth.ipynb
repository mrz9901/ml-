```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Data_Dummys.csv')
data.shape
```

```python
# Function to generate synthetic data
def generate_synthetic_data(data, num_samples):
    synthetic_data = []
    for _ in range(num_samples):
        # Randomly sample an existing row
        row = data.sample(n=1).iloc[0]
        
        # Perturb the values slightly
        synthetic_row = row.copy()
        synthetic_row['Temperatur'] = row['Temperatur'] + np.random.uniform(-2, 2)
        synthetic_row['Kelembaban'] = row['Kelembaban'] + np.random.uniform(-5, 5)
        synthetic_row['Wind speed'] = row['Wind speed'] + np.random.uniform(-10, 10)
        synthetic_row['Ketinggian air'] = row['Ketinggian air'] + np.random.uniform(-5, 5)
        
        # Ensure values stay within realistic bounds
        synthetic_row['Temperatur'] = max(20, min(synthetic_row['Temperatur'], 40))
        synthetic_row['Kelembaban'] = max(0, min(synthetic_row['Kelembaban'], 100))
        synthetic_row['Wind speed'] = max(0, min(synthetic_row['Wind speed'], 100))
        synthetic_row['Ketinggian air'] = max(20, min(synthetic_row['Ketinggian air'], 70))
        
        synthetic_data.append(synthetic_row)
    
    return pd.DataFrame(synthetic_data)
```

```python
## Create X ammount of synthetic data
synthetic_data = generate_synthetic_data(data, 1800)
```

```python
augmented_data = pd.concat([data, synthetic_data], ignore_index=True)
```

```python
augmented_data.to_csv('augmented_dummy_data.csv', index=False)
augmented_data.shape
```
